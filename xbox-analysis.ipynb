### Question 2: Which year had the highest global sales

```python
highest_year = data.groupby('Year')['Global'].sum()
highest_year
```

```python
years = [year for year, df in data.groupby('Year')]
years
```

```python
plt.bar(years, highest_year)
plt.xlabel('Year')
plt.ylabel('Global Sales (million $)')
plt.show()

# Answer: 2015 with 68.61 million sales
```

#### Question 3 What are the most popular genres globally and in each region (North America, Europe, Japan, Rest of the World)?

```python
# global
popular_genres_global = data.groupby('Genre')['Global'].sum()
popular_genres_global
```

```python
# Top 5 most popular genres
genre = [genre for genre, df in data.groupby('Genre')]
genre

plt.bar(genre, popular_genres_global)
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.xticks(rotation='vertical')
plt.show()

# Answer: Shooter, Action, Sports, Racing, Action-Adventure
```

```python
# North America 
popular_genres_america = data.groupby('Genre')['North America'].sum()
popular_genres_america
```

```python
plt.bar(genre, popular_genres_america)
plt.xlabel('Genre')
plt.ylabel('North America Sales')
plt.xticks(rotation='vertical')
plt.show()

# Answer : Shooter, Action, Sports, Action-Adventure, Racing
```
